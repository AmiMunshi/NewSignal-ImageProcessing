Grey level resolution

```python
# Import cv2, matplotlib, numpy
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read the original image and know its type
img1= cv2.imread('/content/cameraman.tif',0)
#type(img1)
np.amax(img1)
```

```python
[m,n]=img1.shape
print(m,n)
```

```python
#Ask for the number of bit in which a pixel has to be represented in the new image
b= int(input("Enter the number of bits for the new image"))
```

```python
#Calculate the constant value to do the mapping
maxold= 256
maxnew= 2**b
constant = (maxnew-1)/(maxold-1)
constant
```

```python

img2= np.zeros((m,n)) #creating new image matrix

#Mapping pixel values in the original image to new image
for i in range(m):
    for j in range(n):
        
        img2[i,j] = constant*img1[i,j]
```

```python

img2= np.round(img2)
```
